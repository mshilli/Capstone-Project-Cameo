All of these indicate how famous a particular person is to the general population. Each of these numeric values will be used to determine the potential price a celebrity should charge for a cameo.

## Test and Train data sets

```python
###Building out Cameo Price Groups so we can use KNN

category =pd.cut(cameo_df['Cameo video price'],bins=[0, 5, 10, 25, 50, 100, 250, 500, 1000, 5000],labels=['1','2','3','4', '5', '6', '7', '8', '9'])
cameo_df.insert(1,'Price Group',category)
cameo_df.head()
```

```python
### Exporting the cameo with price groups to CSV
cameo_df.to_csv('cameo_price_group.csv', index=False)
```

```python
### Getting all the independent variables (aka the X) that are going to be used to predict the cameo price group
X = cameo_df[['Wikipedia', 'FF Celebrity rating',
       'FF Main category rating', 'Catergory_Actors', 'Catergory_Artists',
       'Catergory_Athletes', 'Catergory_Business People', 'Catergory_Chefs',
       'Catergory_Comedians', 'Catergory_Models', 'Catergory_Musicians',
       'Catergory_Personalities', 'Catergory_Public Figures',
       'Catergory_Writers']] .values  #.astype(float)
X[0:5]
```

```python
X
```

```python
### Getting the y aka the dependent variable the cameo price group 
y = cameo_df['Price Group'].values
y[0:5]
```

```python
### data standardization to give the data zero mean and unit variance which is best for culstering 
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
### Splitting the data set into a test and training sets. 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y.astype(str), stratify=y, test_size=0.2)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
#importing what I need for KNN for sklearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
##showing the X train data array 
X_train
```

```python
### looking at KNN with k value of 25
neigh = KNeighborsClassifier(n_neighbors = 25).fit(X_train,y_train)
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
### looking at the accuracy of the train and test set
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python
### looping threw the value of K from 1 to 25 to find the best value of K
Ks = 25
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
### plotting the accuracy of each value of K and the standard deviations to make it a little easier to spot the best value of K

plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

```python
### Elbow Method to find the best value of K
error_rate = []
# Will take some time
for i in range(1,25):
 
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(X_train,y_train)
 pred_i = knn.predict(X_test)
 error_rate.append(np.mean(pred_i != y_test))
```

```python
### Plotting the elbow method to see what value of K has the lowest Error Rate and it makes the elbow method really obvious
plt.figure(figsize=(10,6))
plt.plot(range(1,25),error_rate,color='blue', linestyle='dashed', marker='o',
 markerfacecolor='red', markersize=5)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
